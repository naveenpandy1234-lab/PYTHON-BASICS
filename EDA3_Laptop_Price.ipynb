**Price Distribution**

**Price by Brand**

```python
plt.figure(figsize=(10,5))

sns.boxplot(x='brand', y='price', data=df)

plt.title("Laptop Price by Brand")
plt.xticks(rotation=45)
plt.show()
```

Insight

This shows:

price spread

Example interpretation:

Gaming brands like MSI / Asus may have higher prices.

Answer 1)

The brand with the highest average laptop price is typically MSI (Micro-Star International).

Explanation

MSI is known for high-performance gaming laptops, which include:

powerful CPUs

These features increase the overall price.

```python
df.groupby('brand')['price'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='brand', y='price', data=df)
plt.xticks(rotation=45)
plt.title("Average Laptop Price by Brand")
plt.show()
```

Answer 2)

Yes, gaming laptops generally cost more than student laptops.

Explanation

Gaming laptops require:

powerful CPUs

better cooling systems

Student laptops usually prioritize:

Thus gaming laptops tend to have higher average prices.
